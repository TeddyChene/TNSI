```python
CREATE TABLE equipe (
    numero_equipe integer NOT NULL  ,
    nom_equipe TEXT NOT NULL,
    classement integer ,
    PRIMARY KEY (numero_equipe)
);
```

<ol start="1" style="list-style-type:none;font-family:roman,serif;font-size:1.5em;line-height:1.5em">
    <li>
        <ol start="2">
            <li type="a">Créer la table <code>joueur</code> en exécutant la requête SQL dans la cellule ci-dessous :
                <div class="alert alert-block alert-warning">
<p style="font-family:roman,serif;font-size:0.75em;line-height:1em">Il est plus simple d'ajouter les clés étrangères au moment de la création de la table. L'ordre de création des tables est donc important. Ici, la table <code>equipe</code> n'a pas de contrainte. C'est pour cela qu'on l'a créé en premier.</p></div>
</li>
        </ol>
    </li>
</ol>

```python
CREATE TABLE joueur (
    numero_joueur integer NOT NULL  ,
    numero_equipe integer NOT NULL ,
    pseudo TEXT NOT NULL,
    nom TEXT NOT NULL,
    prenom TEXT NOT NULL,
    PRIMARY KEY (numero_joueur),
    CONSTRAINT fk_equipe_joueur
        FOREIGN KEY (numero_equipe) 
        REFERENCES equipe(numero_equipe)
);
```
